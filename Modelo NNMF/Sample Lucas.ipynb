```python
def stringtolist(text):
    text = text.replace("[","")
    text = text.replace("]","")
    text = text.replace(" ","")
    text = text.replace("'","")
    t = text.split(",")
    t = " ".join(t)
    return t

sample_final['token_filtered'] = sample_final['token'].apply(lambda x: stringtolist(str(x)))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
tfidf_vect = CountVectorizer(strip_accents='unicode', ngram_range=(1,5),max_df=0.80, min_df=5, max_features=600, stop_words=stop_words)
tfidf_matrix = tfidf_vect.fit_transform(sample_final['token_filtered'])
```

```python
tfidf_matrix.shape
```

```python
df_tfidf = pd.DataFrame(tfidf_matrix.todense(), columns = tfidf_vect.get_feature_names())
df_tfidf.head()
```

```python
n_components = 5

nnmf = NMF(n_components=n_components, random_state=42)
c = nnmf.fit_transform(df_tfidf)
```

```python
topics = pd.DataFrame(nnmf.components_, columns=tfidf_vect.get_feature_names()).transpose()
```

```python
df_vectorized = pd.DataFrame(c)
df_vectorized['status_id'] = np.array(sample_final['status_id'])
df_vectorized['full_text'] = np.array(sample_final['text'])
df_vectorized['date'] = np.array(sample_final['date'])
```

```python
import numpy as np
narray = np.arange(0,n_components)

for n in narray:
    print('COMPONENTE:', str(n))
    print('-- \n')
    print('Palavras \n')
    print(topics[n].sort_values(ascending=False).head(15))
    print('-- \n')
    print('Tweets \n')
    t_list = np.array(df_vectorized[[0,'full_text']].sort_values(by=0, ascending=False).head(30)['full_text'])
    for t in t_list:
        print(t)
    print('___________________________________________________________________')
```
